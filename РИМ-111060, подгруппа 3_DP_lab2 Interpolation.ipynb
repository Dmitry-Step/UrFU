### 1.2 –ü–æ–¥–∫–ª—é—á–∏—Ç–µ –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–µ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏

### 1.3 –°–æ–∑–¥–∞–π—Ç–µ —Ñ—É–Ω–∫—Ü–∏—é, –≤–æ–∑–≤—Ä–∞—â–∞—é—â—É—é –º–≥–Ω–æ–≤–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å–∏–≥–Ω–∞–ª–∞ 

$F = A \cdot sin(\omega \cdot t + \varphi_0)$,

–≥–¥–µ:

$A$ = –≤—Ç–æ—Ä–∞—è —Å –∫–æ–Ω—Ü–∞ —Ü–∏—Ñ—Ä–∞ –≤ –Ω–æ–º–µ—Ä–µ —Å—Ç—É–¥–µ–Ω—á–µ—Å–∫–æ–≥–æ (–µ—Å–ª–∏ —ç—Ç–æ 0, —Ç–æ –ª—é–±–∞—è —Ü–∏—Ñ—Ä–∞ –¥–æ –Ω–µ–µ);

$t$ - –æ—Ç—Å—á–µ—Ç—ã –≤—Ä–µ–º–µ–Ω–∏ —Å–∏–≥–Ω–∞–ª–∞;

$\omega = 2 \cdot \pi \cdot N$

$\varphi_0 = \frac{2 \cdot \pi}{N}$

$N$ - –ø–æ—Å–ª–µ–¥–Ω—è—è —Ü–∏—Ñ—Ä–∞ –≤ –Ω–æ–º–µ—Ä–µ —Å—Ç—É–¥–µ–Ω—á–µ—Å–∫–æ–≥–æ –±–∏–ª–µ—Ç–∞, –µ—Å–ª–∏ –æ–Ω–∞ –Ω–æ–ª—å, —Ç–æ –ø–µ—Ä–≤–∞—è –ø–µ—Ä–µ–¥ –Ω–µ–π –Ω–µ —Ä–∞–≤–Ω–∞—è 0

```python
def mySignal(A, Nv, t):
    # A - –∞–º–ø–ª–∏—Ç—É–¥–∞
    # t - –º–æ–º–µ–Ω—Ç—ã –≤—Ä–µ–º–µ–Ω–∏
    # N - –Ω–æ–º–µ—Ä –≤–∞—Ä–∏–∞–Ω—Ç–∞
    w = 2 * np.pi * N
    f0 =2 * np.pi / N 
    s = A * np.sin(w * t + f0)
    return s
```

–ü–æ—Å—Ç—Ä–æ–∏–º –ø–æ–ª—É—á–µ–Ω–Ω—ã–π —Å–∏–≥–Ω–∞–ª, –Ω–µ –∑–∞–±—É–¥—å—Ç–µ –ø–æ–¥–ø–∏—Å–∞—Ç—å –Ω–∞–∑–≤–∞–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ –∏ –æ—Å–∏

```python
# –¥–æ–ø–∏—Å–∞—Ç—å
def InitSignal(time, A, N, func):
    pointsY = []
    for t in time:
        pointsY.append(func(A,N,t))
    return pointsY
A = 1
N = 6

time = np.linspace(0, 1, 100)
F = InitSignal(time, A, N, mySignal)

plt.plot(time, F)
plt.xlabel('t')
plt.ylabel('F')
plt.title('Signal')
```

–ü–æ–ª—É—á–∏–º –¥–≤–∞ –ø–µ—Ä–∏–æ–¥–∞ –Ω–∞—à–µ–≥–æ —Å–∏–≥–Ω–∞–ª–∞ –¥–ª—è –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –≤ —Å–ª–µ–¥—É—é—â–∏—Ö –∑–∞–¥–∞–Ω–≥–∏—è—Ö

```python
t = np.linspace(0-(1/N), (1/N), 1000)
Signal = InitSignal(t, A, N, mySignal)

plt.plot(t, Signal)
plt.xlabel('t')
plt.ylabel('F')
plt.title('Signal —Å –¥–≤—É–º—è –ø–µ—Ä–∏–æ–¥–∞–º–∏')
plt.show()
```

### 2.1 –í—ã–±–µ—Ä–∏—Ç–µ –¥–ª–∏—Ç–µ–ª—å–Ω–æ—Å—Ç—å –≤—Ä–µ–º–µ–Ω–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ $T_s$ —Ç–∞–∫–æ–π, —á—Ç–æ–±—ã –Ω–∞ –Ω–µ–º —É–∫–ª–∞–¥—ã–≤–∞–ª–æ—Å—å 2 –ø–µ—Ä–∏–æ–¥–∞ –∞–Ω–∞–ª–∏–∑–∏—Ä—É–µ–º–æ–≥–æ —Å–∏–≥–Ω–∞–ª–∞. –ó–∞–¥–∞–π—Ç–µ –ø–∞—Ä–∞–º–µ—Ç—Ä —á–∞—Å—Ç–æ—Ç—ã –¥–∏—Å–∫—Ä–µ—Ç–∏–∑–∞—Ü–∏–∏ —Ç–∞–∫, —á—Ç–æ–±—ã –≤ —Ñ—É–Ω–∫—Ü–∏—é –º–æ–∂–Ω–æ –±—ã–ª–æ –ø–µ—Ä–µ–¥–∞–≤–∞—Ç—å —á–∏—Å–ª–æ –æ—Ç—Å—á–µ—Ç–æ–≤.


```python
def myDiscretSignal(A, Nv, Np):  # –¥–æ–ø–∏—Å–∞—Ç—å
    # Np - –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ç–æ—á–µ–∫ –Ω–∞ –æ–¥–∏–Ω –ø–µ—Ä–∏–æ–¥ —Å–∏–≥–Ω–∞–ª–∞
    startpoint = 0 - (1/Nv)
    endpoint = 1/Nv
    count = (abs(startpoint)+abs(endpoint))/(1/Nv)
    t = np.linspace(startpoint, endpoint, int(Np*count))
    s = InitSignal(t, A, Nv, mySignal)
    return t,s

timeTest, s = myDiscretSignal(1, 1, 10)
plt.plot(timeTest,s,marker="o")
plt.xlabel('t')
plt.ylabel('F')
plt.title('Discret signal')
plt.show()
```

## –ó–∞–¥–∞–Ω–∏–µ 3 –ò–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏—è

### 3.1 –°–æ–∑–¥–∞–π—Ç–µ 5 —Ñ—É–Ω–∫—Ü–∏–π, –≤–æ–∑–≤—Ä–∞—â–∞—é—â–∏–µ –∏–Ω—Ç–µ—Ä–ø–æ–ª–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å–∏–≥–Ω–∞–ª–∞:

–õ–∏–Ω–µ–π–Ω–∞—è –∏–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏—è - –∏–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏—è –¥–≤—É—á–ª–µ–Ω–æ–º $P(x) = a \cdot x + b$, –≤ –æ–±—â–µ–º —Å–ª—É—á–∞–µ:

$P(x) = f(x_0) + \frac{f(x_1) - f(x_0)}{x_1 - x_0} \cdot (x - x_0)$

[–ë–æ–ª—å—à–µ –∏–Ω—Ñ–æ—Ä–∞–º—Ü–∏–∏](https://ru.wikipedia.org/wiki/–õ–∏–Ω–µ–π–Ω–∞—è_–∏–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏—è)

–ò–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏—è –º–Ω–æ–≥–æ—á–ª–µ–Ω–æ–º –º–Ω–æ–≥–æ—á–ª–µ–Ω–∞ N-–æ–π —Å—Ç–µ–ø–µ–Ω–∏, —Å–æ–≥–ª–∞—Å–Ω–æ —Ñ–æ—Ä–º—É–ª–µ:

$P_{n}(x_{i})=a_{0}+a_{1}x_{i}+a_{2}x_{i}^{2}+\ldots +a_{n}x_{i}^{n}=f(x_{i}),\quad i=0, 1, \ldots , n.$

[–ë–æ–ª—å—à–µ –∏–Ω—Ñ–æ—Ä–∞–º—Ü–∏–∏](https://ru.wikipedia.org/wiki/–ò–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏—è_–∞–ª–≥–µ–±—Ä–∞–∏—á–µ—Å–∫–∏–º–∏_–º–Ω–æ–≥–æ—á–ª–µ–Ω–∞–º–∏)

–ò–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏—è –º–Ω–æ–≥–æ—á–ª–µ–Ω–æ–º –ù—å—é—Ç–æ–Ω–∞ —Å–æ–≥–ª–∞—Å–Ω–æ –≤—ã—Ä–∞–∂–µ–Ω–∏—é:

$P_n(x) = y_0 + q \Delta y_0 + \frac{q(q-1)}{2!} \Delta^2 y_0 + \ldots + \frac{q(q-1)\ldots (q-n+1)}{n!} \Delta^n y_0$, –≥–¥–µ $q = \frac{x-x_0}{h}, y_i = f_i$

[–ë–æ–ª—å—à–µ –∏–Ω—Ñ–æ—Ä–∞–º—Ü–∏–∏](https://ru.wikipedia.org/wiki/–ò–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏–æ–Ω–Ω—ã–µ_—Ñ–æ—Ä–º—É–ª—ã_–ù—å—é—Ç–æ–Ω–∞)

–ò–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏—è –º–Ω–æ–≥–æ—á–ª–µ–Ω–æ–º –õ–∞–≥—Ä–∞–Ω–∂–∞, —Å–æ–≥–ª–∞—Å–Ω–æ –≤—ã—Ä–∞–∂–µ–Ω–∏—é

$L(x)=\sum _{i=0}^{n}y_{i}l_{i}(x)$

$l_{i}(x) = \prod_{j=0,j \neq i}^{n} \frac{x-x_j}{x_i-x_j} = \frac{x-x_0}{x_i-x_0}  \ldots \frac{x-x_{i-1}}{x_i-x_{i-1}} \cdot \frac{x-x_{i+1}}{x_i-x_{i+1}} \ldots \frac{x-x_n}{x_i - x_n}$

[–ë–æ–ª—å—à–µ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏](https://ru.wikipedia.org/wiki/–ò–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏–æ–Ω–Ω—ã–π_–º–Ω–æ–≥–æ—á–ª–µ–Ω_–õ–∞–≥—Ä–∞–Ω–∂–∞)

$S_i(x) = a_i + b_i(x-x_i) + c_i(x-x_i)^2 + d_i(x-x_i)^3$

```python
def myLinearInterp(x, y):
    x1 = []
    y1 = []
    
    for i in range(len(x)-1):
        newX = (x[i] + x[i+1])/2
        x1.append(x[i])
        x1.append(newX)
        newY = y[i] + ((y[i+1] - y[i])/(x[i+1]-x[i])) * (newX-x[i])
        y1.append(y[i])
        y1.append(newY)
    x1.append(x[len(x)-1])
    y1.append(y[len(y)-1])
    return x1, y1
        
Time1=np.linspace(-1, 1, 100)
Signal1=InitSignal(Time1, A, N, mySignal)
timeDis1, signalDis1 = myDiscretSignal(2, 1, 4)
timeInt1, signalInt1 = myLinearInterp(timeDis1, signalDis1)
print(len(timeInt1))
timeInt1, signalInt1 = myLinearInterp(timeInt1, signalInt1)
print(len(timeInt1))
timeInt1, signalInt1 = myLinearInterp(timeInt1, signalInt1)
print(len(timeInt1))

b = interp1d(timeDis1, signalDis1, kind='linear')
y_linear = b(newX)

fig = plt.figure(figsize=(20, 15), dpi=100)
plt.subplot(3, 1, 1)
plt.plot(Time1, Signal1, color='blue', label='–ê–Ω–∞–ª–æ–≥–æ–≤—ã–π —Å–∏–≥–Ω–∞–ª')
plt.plot(timeDis1, signalDis1, color='green', label='–î–∏—Å–∫—Ä–µ—Ç–Ω—ã–π —Å–∏–≥–Ω–∞–ª')
plt.plot(timeInt1, signalInt1, color='red', label='–°–∏–≥–Ω–∞–ª –ø–æ –∞–ª–≥–æ—Ä–∏—Ç–º—É –∏–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏–∏')
plt.plot(newX, y_linear, color='magenta', label='–ò–Ω—Ç–µ—Ä–ø–æ–ª. —Å–∏–≥–Ω–∞–ª –≤—Å—Ç—Ä. –±–∏–±–ª.')
plt.title('–õ–∏–Ω–µ–π–Ω–∞—è –∏–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏—è')
plt.grid()
plt.legend()

def myPolynomInterp(x, y):
    m = []
    for j in range (0,len(x)):
        s = []
        for k in range(0,len(x)):
            s.append(x[j]**k)
        m.append(s)
    a = np.linalg.solve(m,y)
    x1 = []
    y1 = []
    for i in range(0,len(x)-1):
        x1.append(x[i])
        y1.append(y[i])
        newX = (x[i] + x[i+1])/2
        x1.append(newX)
        newY = 0
        for j in range(0,len(a)):
            newY = newY + newX**j*a[j]
        y1.append(newY)
    x1.append(x[len(x)-1])
    y1.append(y[len(y)-1])
    return x1, y1

Time2=np.linspace(-1, 1, 100)
Signal2=InitSignal(Time2, A, N, mySignal)
timeDis2, signalDis2 = myDiscretSignal(2, 1, 4)
timeInt2, signalInt2 = myPolynomInterp(timeDis2, signalDis2)
print(len(timeInt2))
timeInt2, signalInt2 = myPolynomInterp(timeInt2, signalInt2)
print(len(timeInt2))
timeInt2, signalInt2 = myPolynomInterp(timeInt2, signalInt2)
print(len(timeInt2))

initmodel = np.poly1d(np.polyfit(timeDis2, signalDis2,7))
y_polynom = initmodel(newX)

fig = plt.figure(figsize=(20, 15), dpi=100)
plt.subplot(3, 1, 2)
plt.plot(Time2, Signal2, color='green', label='–ê–Ω–∞–ª–æ–≥–æ–≤—ã–π —Å–∏–≥–Ω–∞–ª')
plt.plot(timeDis2, signalDis2, color='black', label='–î–∏—Å–∫—Ä–µ—Ç–Ω—ã–π —Å–∏–≥–Ω–∞–ª')
plt.plot(timeInt2, signalInt2, color='blue', label='–°–∏–≥–Ω–∞–ª –ø–æ –∞–ª–≥–æ—Ä–∏—Ç–º—É –∏–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏–∏')
plt.plot(newX, y_polynom, color='yellow', label='–ò–Ω—Ç–µ—Ä–ø–æ–ª. —Å–∏–≥–Ω–∞–ª –≤—Å—Ç—Ä. –±–∏–±–ª.')
plt.title('–ò–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏—è –º–Ω–æ–≥–æ—á–ª–µ–Ω–æ–º')
plt.grid()
plt.legend()

def Lagrange(x, y, t):
    z=0
    for j in range(len(y)):
        p1=1
        p2=1
        for i in range(len(x)):
            if i!=j:
                p1 = p1*(t-x[i])
                p2 = p2*(x[j]-x[i])
        z = z+y[j]*p1/p2
    return z

def myLagrangeInterp(x, y):
    newX = []
    newY = []
    for i in range(0,len(x)-1):
        newX.append(x[i])
        newY.append(y[i])
        middleX = (x[i] + x[i+1])/2
        middleY = Lagrange(x, y, middleX)
        newX.append(middleX)
        newY.append(middleY)
    newX.append(x[len(x)-1])
    newY.append(y[len(y)-1])
    return newX, newY

Time3=np.linspace(-1, 1, 100)
Signal3=InitSignal(Time3, A, N, mySignal)
timeDis3, signalDis3 = myDiscretSignal(2, 1, 4)
timeInt3, signalInt3 = myLagrangeInterp(timeDis3, signalDis3)
print(len(timeInt3))
timeInt3, signalInt3 = myLagrangeInterp(timeInt3, signalInt3)
print(len(timeInt3))
timeInt3, signalInt3 = myLagrangeInterp(timeInt3, signalInt3)
print(len(timeInt3))

lpoly = lagrange(timeDis3, signalDis3)
y_lagrange = lpoly(newX)

fig = plt.figure(figsize=(20, 15), dpi=100)
plt.subplot(3, 1, 3)
plt.plot(Time3, Signal3, color='green', label='–ê–Ω–∞–ª–æ–≥–æ–≤—ã–π —Å–∏–≥–Ω–∞–ª')
plt.plot(timeDis3, signalDis3, color='blue', label='–î–∏—Å–∫—Ä–µ—Ç–Ω—ã–π —Å–∏–≥–Ω–∞–ª')
plt.plot(timeInt3, signalInt3, color='magenta', label='–°–∏–≥–Ω–∞–ª –ø–æ –∞–ª–≥–æ—Ä–∏—Ç–º—É –∏–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏–∏')
plt.plot(newX, y_lagrange, color='red', label='–ò–Ω—Ç–µ—Ä–ø–æ–ª. —Å–∏–≥–Ω–∞–ª –≤—Å—Ç—Ä. –±–∏–±–ª.')
plt.title('–ò–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏—è –º–Ω. –õ–∞–≥—Ä–∞–Ω–∂–∞')
plt.grid()
plt.legend()

plt.show()
```

–î–∞–ª–µ–µ, —Ä–∞—Å—Å–º–æ—Ç—Ä–∏–º –∏–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏—é –∫—É–±–∏—á–µ—Å–∫–∏–º —Å–ø–ª–∞–π–Ω–æ–º, —Å–æ–≥–ª–∞—Å–Ω–æ –≤—ã—Ä–∞–∂–µ–Ω–∏—é:
$S_i(x) = a_i + b_i(x-x_i) + c_i(x-x_i)^2 + d_i(x-x_i)^3$

```python
def cubic_interp1d(x0, x, y):
    
    x = np.asfarray(x)
    y = np.asfarray(y)

    # remove non finite values
    # indexes = np.isfinite(x)
    # x = x[indexes]
    # y = y[indexes]

    # check if sorted
    if np.any(np.diff(x) < 0):
        indexes = np.argsort(x)
        x = x[indexes]
        y = y[indexes]

    size = len(x)

    xdiff = np.diff(x)
    ydiff = np.diff(y)

    # allocate buffer matrices
    Li = np.empty(size)
    Li_1 = np.empty(size-1)
    z = np.empty(size)

    # fill diagonals Li and Li-1 and solve [L][y] = [B]
    Li[0] = sqrt(2*xdiff[0])
    Li_1[0] = 0.0
    B0 = 0.0 # natural boundary
    z[0] = B0 / Li[0]

    for i in range(1, size-1, 1):
        Li_1[i] = xdiff[i-1] / Li[i-1]
        Li[i] = sqrt(2*(xdiff[i-1]+xdiff[i]) - Li_1[i-1] * Li_1[i-1])
        Bi = 6*(ydiff[i]/xdiff[i] - ydiff[i-1]/xdiff[i-1])
        z[i] = (Bi - Li_1[i-1]*z[i-1])/Li[i]

    i = size - 1
    Li_1[i-1] = xdiff[-1] / Li[i-1]
    Li[i] = sqrt(2*xdiff[-1] - Li_1[i-1] * Li_1[i-1])
    Bi = 0.0 # natural boundary
    z[i] = (Bi - Li_1[i-1]*z[i-1])/Li[i]

    # solve [L.T][x] = [y]
    i = size-1
    z[i] = z[i] / Li[i]
    for i in range(size-2, -1, -1):
        z[i] = (z[i] - Li_1[i-1]*z[i+1])/Li[i]

    # find index
    index = x.searchsorted(x0)
    np.clip(index, 1, size-1, index)

    xi1, xi0 = x[index], x[index-1]
    yi1, yi0 = y[index], y[index-1]
    zi1, zi0 = z[index], z[index-1]
    hi1 = xi1 - xi0

    # calculate cubic
    f0 = zi0/(6*hi1)*(xi1-x0)**3 + \
         zi1/(6*hi1)*(x0-xi0)**3 + \
         (yi1/hi1 - zi1*hi1/6)*(x0-xi0) + \
         (yi0/hi1 - zi0*hi1/6)*(xi1-x0)
    return f0

if __name__ == '__main__':
    import matplotlib.pyplot as plt
    x = np.linspace(0, 10, 11)
    y = np.sin(x)
    plt.scatter(x, y)

    x_new = np.linspace(0, 10, 201)
    plt.plot(x_new, cubic_interp1d(x_new, x, y))

    plt.show()
```

### 3.5 –ò—Å—Å–ª–µ–¥—É–π—Ç–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–µ —Å–≤–æ–π—Å—Ç–≤–∞ —Å–≥–Ω–∞–ª–∞ –¥–æ –∏–Ω—Ç–µ—Ä–ø–æ–ª—è—Ü–∏–∏ –∏ –ø–æ—Å–ª–µ, –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ –≤ –ø—Ä–µ–¥—ã–¥—É—â–µ–º –∑–∞–¥–∞–Ω–∏–∏.

$m = \frac{1}{N}\sum_{n=0}^{N-1}x(n)$ - –ú–û

$sigma^2 = \frac{1}{N-1}\sum_{n=0}^{N-1}|x(n) - m|^{2}$ - –¥–∏—Å–ø–µ—Ä—Å–∏—è



```python
# –∫–æ–¥
OurTime1 = np.linspace(-1, 1, 1024)
OurSignal1 = InitSignal(OurTime1, A, N, mySignal)
timeDis1, SignalDis1 = myDiscretSignal(2, 1, 4)

b = interp1d(timeDis1, signalDis1, kind='linear')
y_linear = b(OurTime1)

initmodel = np.poly1d(np.polyfit(timeDis2, signalDis2,7))
y_polynom = initmodel(OurTime1)

lpoly = lagrange(timeDis3, signalDis3)
y_lagrange = lpoly(OurTime1)

def findMX(y):
    sum=0
    for i in y:
        sum += 1
    return sum/len(y)

def findDX(y,m):
    sum=0
    for i in y:
        sum += abs(i-m)**2
    return sum/len(y)


MX1 = findMX(OurSignal1)
MXnp1 = np.mean(OurSignal1)
print('MO –ø–æ —Ñ–æ—Ä–º—É–ª–µ (–ù–∞—à —Å–∏–≥–Ω–∞–ª) = ', MX1, ' MO –≤—Å—Ç—Ä–æ–µ–Ω–Ω–∞—è (–ù–∞—à —Å–∏–≥–Ω–∞–ª) = ', MXnp1)

MX2 = findMX(y_linear)
MXnp2 = np.mean(y_linear)
print('MO –ø–æ —Ñ–æ—Ä–º—É–ª–µ (–ò–Ω—Ç. –ª–∏–Ω.) = ', MX2, ' MO –≤—Å—Ç—Ä–æ–µ–Ω–Ω–∞—è (–ò–Ω—Ç. –ª–∏–Ω.) = ', MXnp2)

MX3 = findMX(y_polynom)
MXnp3 = np.mean(y_polynom)
print('MO –ø–æ —Ñ–æ—Ä–º—É–ª–µ (–ò–Ω—Ç. –º–Ω–æ–≥–æ—á–ª.) = ', MX3, ' MO –≤—Å—Ç—Ä–æ–µ–Ω–Ω–∞—è (–ò–Ω—Ç. –º–Ω–æ–≥–æ—á–ª.) = ', MXnp3)

MX4 = findMX(y_lagrange)
MXnp4 = np.mean(y_lagrange)
print('MO –ø–æ —Ñ–æ—Ä–º—É–ª–µ (–ò–Ω—Ç. –õ–∞–≥—Ä.) = ', MX4, ' MO –≤—Å—Ç—Ä–æ–µ–Ω–Ω–∞—è (–ò–Ω—Ç. –õ–∞–≥—Ä.) = ', MXnp4)

DX1 = findDX(OurSignal1, MX1)
DXnp1 = np.var(OurSignal1)
print('DX –ø–æ —Ñ–æ—Ä–º—É–ª–µ (–ù–∞—à —Å–∏–≥–Ω–∞–ª) = ', DX1, ' DX –≤—Å—Ç—Ä–æ–µ–Ω–Ω–∞—è (–ù–∞—à —Å–∏–≥–Ω–∞–ª) = ', DXnp1)

DX2 = findDX(y_linear, MX2)
DXnp2 = np.var(y_linear)
print('DX –ø–æ —Ñ–æ—Ä–º—É–ª–µ (–ò–Ω—Ç. –ª–∏–Ω.) = ', DX2, ' DX –≤—Å—Ç—Ä–æ–µ–Ω–Ω–∞—è (–ò–Ω—Ç. –ª–∏–Ω.) = ', DXnp2)

DX3 = findDX(y_polynom, MX3)
DXnp3 = np.var(y_polynom)
print('DX –ø–æ —Ñ–æ—Ä–º—É–ª–µ (–ò–Ω—Ç. –º–Ω–æ–≥–æ—á–ª.) = ', DX3, ' DX –≤—Å—Ç—Ä–æ–µ–Ω–Ω–∞—è (–ò–Ω—Ç. –º–Ω–æ–≥–æ—á–ª.) = ', DXnp3)

DX4 = findDX(y_lagrange, MX4)
DXnp4 = np.var(y_lagrange)
print('DX –ø–æ —Ñ–æ—Ä–º—É–ª–µ (–ò–Ω—Ç. –õ–∞–≥—Ä.) = ', DX4, ' DX –≤—Å—Ç—Ä–æ–µ–Ω–Ω–∞—è (–ò–Ω—Ç. –õ–∞–≥—Ä.) = ', DXnp4)
```
