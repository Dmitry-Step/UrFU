```python
# importing matplotlib module
from matplotlib import pyplot as plt
 
# x-axis values
x = [5, 2, 9, 4, 7]
 
# Y-axis values
y = [10, 5, 8, 4, 2]
 
# Function to plot
plt.plot(x, y)
 
# function to show the plot
plt.show()
```

```python
x = list(map(lambda x: x * 0.1, range(-50,50)))
x[3]
```

```python
y = list(map(lambda v: v**2, x))
y[3]
```

```python
# Function to plot
plt.plot(x, y)
 
# function to show the plot
plt.show()
```

```python
import math as m
x = [i * 0.1 for i in range(-50,50)]
y = [m.sin(i) for i in x]
```

```python
# Function to plot
plt.plot(x, y)
 
# function to show the plot
plt.show()
```
